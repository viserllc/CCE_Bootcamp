```python
#CREATING SCATTER GRAPH
import matplotlib.pyplot as plt

height = [120, 125, 128,135,150,135]
weight = [70, 68, 75, 90, 98, 100]

# plotting the points  
#alpha = 0 (plots fully transparent and unable to view on a white background)
# alpha = 1 (default value with non-transparent points)

plt.scatter(height, weight, color = 'red', s=75, alpha = 0.8) 
  
# naming the x axis 
plt.xlabel('height') 
# naming the y axis 
plt.ylabel('weight') 
  
# giving a title to my graph 
plt.title('My first graph!') 
  
# function to show the plot 
plt.show();
```

```python

height = [120, 125, 128,135,150,135]
weight = [70, 68, 75, 90, 98, 100]

# plotting the points  
plt.plot(weight,height,ls = "dotted", lw = 3) 
  
# naming the x axis 
plt.xlabel('height') 
# naming the y axis 
plt.ylabel('weight') 

 
#giving a title to the graph
plt.title('My first graph!') 
  
# function to show the plot 
plt.show()
```

```python
# Another type of line plot......

# x axis values 
x = [1,2,3,4,5,6] 
# corresponding y axis values 
y = [2,4,1,5,2,6] 
  
# plotting the points  
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=12) 

# naming the x axis 
plt.xlabel('x-values') 
# naming the y axis 
plt.ylabel('y-values') 

 
#giving a title to the graph
plt.title('My first graph!') 
  
# function to show the plot 
plt.show()
  
```

```python
  
# x-coordinates of bars  
left = [1, 2, 3, 4, 5] 
  
# heights of bars 
height = [10, 24, 36, 40, 5] 
  
# labels for bars 
tick_label = ['one', 'two', 'three', 'four', 'five'] 
  
# plotting a bar chart 
plt.bar(left, height, tick_label = tick_label, width = 0.8, color = ['red', 'green']) 
  
# naming the x-axis 
plt.xlabel('x - axis') 
# naming the y-axis 
plt.ylabel('height') 
# plot title 
plt.title('Bar chart!') 
  
# function to show the plot ---- SOMETIMES ";" IS NEEDED AFTER PLT.SHOW() TO DISPLAY THE GRAPH
plt.show();
```

```python
#CREATING HISTOGRAM
ages = [2,5,70,40,30,45,50,45,43,40,44,60,7,13,57,18,90,77,32,21,20,40] 
  
# setting the ranges and no. of intervals 
range = (0, 100) 
bins = 10  
  
# plotting a histogram 
plt.hist(ages, bins, range, color = 'green', rwidth = 0.8) 
  
# x-axis label 
plt.xlabel('age') 
# frequency label 
plt.ylabel('No. of people') 
# plot title 
plt.title('My histogram') 
  
# function to show the plot 
plt.show()
```

```python
# CREATING PIE CHART
#More details about the parameters of plt.pie can be found at: 
#"https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.pie.html"

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['pink', 'lightgreen', 'gold', 'skyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
#
plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, startangle=90)

plt.axis("equal")

plt.show()
```
